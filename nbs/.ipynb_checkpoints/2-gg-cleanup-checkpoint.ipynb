```python
# read in CSV
df_raw = pd.read_csv("lit-canon-KEY.csv")
```

```python
# HOUSEKEEPING
df = df_raw.copy()

df = df.drop(185)
df = df.reset_index(drop=True) # resetting index after dropping stupid row
```

## 1. Cleaning up prolematic values: years + cover_ids

### YEAR MISSING

```python
# filtering based on NaN in year column
no_years = df[df['year'].isna()].copy()
no_years
```

```python
# rewriting by index + cold
df_idxs = [143, 316, 327]
years = [1842, 1954, 2003]


for i in range(len(years)):
    year_replaced = years[i]
    idx = df_idxs[i]
    df.year.iloc[idx] = year_replaced
    
df.iloc[[143, 316, 327]]
```

### CONVERTING FROM FLOATS TO INTEGER TYPES

```python
# type conversion +  dealing with NaN
# https://pandas.pydata.org/docs/user_guide/missing_data.html

df[['year', 'cover_id']] = df[['year', 'cover_id']].fillna(0.0).astype(int)
df[['year', 'cover_id']].head()
```

## 2. Cleaning up problematic values: authors

### NO AUTHORS LISTED

```python
# filtering for no author in author column

no_authors = df[df['author'].str.contains('No author listed')].copy()
no_authors
```

```python
# permanently rewriting + checking

df.author.iloc[316] = 'Friedrich Nietzsche'
df.iloc[316]
```

## 3. Cleaning up problematic values: titles or author not in english

```python
# get list of weird authors

## helper function\

def isEnglish(s):
    try:
        s.encode(encoding='utf-8').decode('ascii')
    except UnicodeDecodeError:
        return False
    else:
        return True

df_eng = df[['author', 'title']].copy()
```

```python
# getting count of values not encoded in english
## https://stackoverflow.com/questions/27084617/detect-strings-with-non-english-characters-in-python
df_bools = df_eng.applymap(isEnglish)
df_bools

# need to locate elements where encoding is wonky
false_authors = df_bools.loc()[(~df_bools.author)].copy()
false_authors
# false_authors.value_counts()

# grabbing the indeces
false_authors_idx = false_authors.index.values.tolist()
false_authors_idx
```

```python
# checking for title
## https://stackoverflow.com/questions/18358938/get-row-index-values-of-pandas-dataframe-as-list

false_titles = df_bools.loc()[(~df_bools.title)]
false_titles
# false_titles.value_counts()

# grabbing the indeces
false_titles_idx = false_titles.index.values.tolist()
false_titles_idx
```

```python
# create new DF based on index list of false authors and titles
## https://stackoverflow.com/questions/1319338/combining-two-lists-and-removing-duplicates-without-removing-duplicates-in-orig

in_authors = set(false_authors_idx)
in_titles = set(false_titles_idx)

in_second_but_not_in_first = in_titles - in_authors

merged_idx = list(in_authors) + list(in_second_but_not_in_first)
merged_idx.sort()
```

```python
# creating new df based on indeces

weird_encoding = df_eng.iloc[merged_idx].copy()
weird_encoding
```
