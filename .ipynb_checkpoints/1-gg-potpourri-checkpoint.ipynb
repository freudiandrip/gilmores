```python
import requests
import pprint as pp
import re
import time
from tqdm import tqdm
from multiprocessing.pool import ThreadPool
import csv
import json
```

```python
#API call 1 - currently reading shelf

url_currently = "https://openlibrary.org/people/kim_nguyen377/books/currently-reading.json"

response_currently = requests.get(url_currently)

# Request was successful
if response_currently.status_code == 200:
    print("Status success!")
    # Extract JSON data from the response
    data_currently = response_currently.json()
    # pp.pprint(data_currently)

# Request failed
else:
    print("Status error:", response_currently.status_code)
```

```python
#API call 2 - want to read shelf

url_wanna = "https://openlibrary.org/people/kim_nguyen377/books/want-to-read.json"

response_wanna = requests.get(url_wanna)

# Request was successful
if response_wanna.status_code == 200:
    print("Status success!")
    # Extract JSON data from the response
    data_wanna = response_wanna.json()
    # pp.pprint(data_wanna)

# Request failed
else:
    print("Status error:", response_wanna.status_code)
    
```

```python
currently_reading = data_currently['reading_log_entries']
# print(len(currently_reading))

wanna_read = data_wanna['reading_log_entries']
# print(len(wanna_read))

canon_list = currently_reading + wanna_read
# print(len(canon_list))
# pp.pprint(canon_list)
```

```python
## IGNORING - IMPACT (see documentation)

book_list = canon_list
# pp.pprint(book_list)

canon = []

for item in tqdm(book_list):
    # initializing + resetting vars
    work = {}
    book = item['work']
    
    work['title'] = book['title']
    # print("\n---------------------------------------------")
    # print(book['author_names'])
    try: 
        work['author'] = book['author_names'][0]
    
    except: 
        work['author'] = "No author listed"
    
    work['year'] = book['first_publish_year']
    
    # Storing edition stuff as a fail-safe
    work['OLID_work_key'] = book['key']
    work['OLID_logged_ed'] = item['logged_edition']
    
    
    # using covers API
    cover_id = book['cover_id']
    work['cover_id'] = cover_id
    work['cover_sm'] = f"https://covers.openlibrary.org/b/id/{cover_id}-S.jpg"
    work['cover_med'] = f"https://covers.openlibrary.org/b/id/{cover_id}-M.jpg"
    # print("\n---------------------------------------------")
    # pp.pprint(work)
    canon.append(work)
    
# print("\n---------------------------------------------")
# print("compiling clean dict")
# print("---------------------------------------------")
# pp.pprint(canon)
```

```python
# Take all open lib IDs --> search for subjects concurrently in the works page.
url_list = []

for work in canon:
    work_key = work['OLID_work_key']
    url = f"https://openlibrary.org/{work_key}.json"
    url_list.append(url)

def grab_pages(url):
    subjects = []
    r = requests.get(url)
    work = r.json()
    
    try:
        subjects = work['subjects']
        subjects_list.append(subjects)
        
    except:
        subjects = [f"No subjects field found"]
        subjects_list.append(subjects)
```

```python
# SEQUENTIAL SCRAPING
subjects_list = []
subjects = []

for url in tqdm(url_list):
    grab_pages(url)

# since subjects and canon list are 1:1, write new key
for i in tqdm(range(len(subjects_list))):
    canon[i]['subjects'] = subjects_list[i]
    
# pp.pprint(canon)
```

```python
# CONCURRENT SCRAPING - adapted from Kai
# ISSUE: CONCURRENT SCRAPING IS OBTAINING ITEMS OUT OF ORDER
## see: https://stackoverflow.com/questions/16288102/is-it-possible-to-receive-out-of-order-responses-with-http

subjects_list = []
subjects = []

def set_up_threads(function, list):
    
    with ThreadPool() as pool:
        results = []
        
        for result in tqdm(pool.imap(function, list), total=len(list), mininterval=0.5):
            
            try:
                results.append(result)
                
            except Exception as e:
                results.append(f"Failed to scrape {item}: {e}")
                
            time.sleep(0.1) # Add a 100ms delay to slow down the loop return results
         # sanity checking   
        return results
    
set_up_threads(grab_pages, url_list[0:50])
```

```python
# https://stackoverflow.com/questions/10373247/how-do-i-write-a-python-dictionary-to-a-csv-file
# write to csv for processing

pp.pprint(canon)

with open('lit-canon.csv', 'w') as f:  # You will need 'wb' mode in Python 2.x
    w = csv.DictWriter(f, canon[0].keys())
    w.writeheader()
    
    for book in tqdm(canon):
        w.writerow(book)
```
